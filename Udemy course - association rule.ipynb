```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
result = list(rules)
result
```

```python
def inspect(result):
    lhs = [tuple(res[2][0][0])[0] for res in result]
    rhs = [tuple(res[2][0][1])[0] for res in result]
    support = [res[1] for res in result]
    confidence = [res[2][0][2] for res in result]
    lift = [res[2][0][3] for res in result]
    return list(zip (lhs, rhs, support, confidence, lift))
resindf = pd.DataFrame(inspect(result), columns = ['left hand side','right hand side','support','confidence','lift'])
```

```python
resindf
```

```python
resindf.nlargest(n = 10, columns = 'lift')
```
